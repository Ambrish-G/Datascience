```python
"""1) Write a function in python that inputs a dataframe and identify which columns
      have date in them. Using these date columns make new columns which are difference
      between these columns taking 2 at a time. (for instance if there is date1, date2, 
      date3 columns, output should be like date1-date2, date2-date3, date1-date3).
"""


import pandas as pd
from dateutil import parser
import itertools as it
from datetime import datetime


#creating the test data
df = pd.DataFrame({'0': ['a', 'b', 'c'], 
                   '1': ['2015-2-5','2016-March-28', '2015/April/29'], 
                   '2': ['30-1-2015','2016/2/28', '2015/April/29,'],
                   '3': ['2015-1-27','2015-2-28', '2015-3-29']})
df.to_csv('test1.csv', index=False)

#loading the test data
df = pd.read_csv('test1.csv', parse_dates=True)

#data type assign from object to  datetime
df = df.apply(lambda col: pd.to_datetime(col, errors='ignore') 
              if col.dtypes == object 
              else col, 
              axis=0)

#grouping of similar datetime datatypes together
g = df.columns.to_series().groupby(df.dtypes).groups
mylist = list(df.select_dtypes(include=['<M8[ns]']).columns)
print(mylist)

#change all date format to year/month/date of all column

for i in range(len(mylist)) : 
    try:
        df[i]= pd.to_datetime(df[mylist[i]], format='%Y-%m-%d', errors='raise')
        print(df[i])
    except ValueError:
        pass
print()     

#date diference function
def date_diff(i,fd,ld):
    df[i] = (df[fd] - df[ld]).dt.days
    print(df[i])   


#combination principal for group of two dates together
a=list(it.combinations(mylist, 2))
for j in range(len(a)) :  
    for k in range(len(a[j])) :  
        print(a[j][k], end=" ")     
    print() 
    date_diff(j+1,a[j][k-1],a[j][k]) 
    print() 
    

```
